## DNN 분류

```python
from keras.utils import to_categorical

train_cqt_np = np.array(train_audio_cqt, np.float32)
train_cqt_np = train_cqt_np.reshape((80, 168*16))
test_cqt_np = np.array(test_audio_cqt, np.float32)
test_cqt_np = test_cqt_np.reshape((20, 168*16))

train_cqt_array = np.expand_dims(train_cqt_np, -1)
test_cqt_array = np.expand_dims(test_cqt_np, -1)

train_labels_cat = to_categorical(train_labels_np)
test_labels_cat = to_categorical(test_labels_np)

print(train_cqt_array.shape)
print(train_labels_cat.shape)
print(test_cqt_array.shape)
print(test_labels_cat.shape)
```

```python
from keras.models import Sequential, Model
from keras.layers import Input, Dense

def model_build():
    model = Sequential()
    
    input = Input(shape=(2688, ), name='input')
    output = Dense(512, activation='relu', name='hidden1')(input)
    output = Dense(256, activation='relu', name='hidden2')(output)
    output = Dense(128, activation='relu', name='hidden3')(output)
    output = Dense(2, activation="softmax", name='output')(output)
    
    model = Model(inputs=[input], outputs=output)
    
    model.compile(optimizer='adam',
                 loss='categorical_crossentropy',
                 metrics=['acc'])
    return model
```

```python
model = model_build()
model.summary()
```

```python
history = model.fit(train_cqt_array, train_labels_cat, epochs=30, batch_size=24, validation_split=0.2)
```

```python
def plot_history(history_dict):
    loss = history_dict['loss']
    val_loss = history_dict['val_loss']
    
    epochs = range(1, len(loss)+1)
    fig = plt.figure(figsize=(14, 5))
    ax1 = fig.add_subplot(1,2,1)
    ax1.plot(epochs, loss, 'b--', label='train_loss')
    ax1.plot(epochs, val_loss, 'r', label='val_loss')
    ax1.set_xlabel("Epochs")
    ax1.set_ylabel("Loss")
    ax1.grid()
    ax1.legend()
    
    acc = history_dict['acc']
    val_acc = history_dict['val_acc']
    ax2 = fig.add_subplot(1,2,2)
    ax2.plot(epochs, acc, 'b--', label='train_accuracy')
    ax2.plot(epochs, val_acc, 'r', label='val_accuracy')
    ax2.set_xlabel("Epochs")
    ax2.set_ylabel("accuarcy")
    ax2.grid()
    ax2.legend()
    
    plt.show()
```

```python
plot_history(history.history)
```

```python
model.evaluate(test_cqt_array, test_labels_cat)
```

```python
pred = model(test_cqt_array)
## classification_report
from sklearn.metrics import classification_report
y_test = test_labels_np
preds_list = pred.numpy().argmax(-1)
clf_report = classification_report(y_test, preds_list)
print(clf_report)

# 오차행렬 생성
from sklearn.metrics import classification_report, confusion_matrix
cf_matrix = confusion_matrix(y_test, preds_list)
# 오차행렬 시각화
import matplotlib.pyplot as plt
import seaborn as sns
sns.heatmap(cf_matrix, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predict')
plt.ylabel('Actual')
plt.show()
```

## CNN 모델

```python
from keras.utils import to_categorical

train_cqt_np = np.array(train_audio_cqt, np.float32)
#train_cqt_np = train_cqt_np.reshape((80, 168*16))
test_cqt_np = np.array(test_audio_cqt, np.float32)
#test_cqt_np = test_cqt_np.reshape((20, 168*16))

train_cqt_array = np.expand_dims(train_cqt_np, -1)
test_cqt_array = np.expand_dims(test_cqt_np, -1)

train_labels_cat = to_categorical(train_labels_np)
test_labels_cat = to_categorical(test_labels_np)

print(train_cqt_array.shape)
print(train_labels_cat.shape)
print(test_cqt_array.shape)
print(test_labels_cat.shape)
```

```python
from keras.layers import Conv2D, MaxPool2D, Flatten

def CNN_model_build():
    model = Sequential()
    
    input = Input(shape=(168, 16, 1))
    output = Conv2D(128, 3, strides=1, padding='same', activation='relu')(input)
    output = MaxPool2D(pool_size=(2,2), strides=2, padding='same')(output)
    
    output = Conv2D(256, 3, strides=1, padding='same', activation='relu')(output)
    output = MaxPool2D(pool_size=(2,2), strides=2, padding='same')(output)
    
    output = Conv2D(512, 3, strides=1, padding='same', activation='relu')(output)
    output = MaxPool2D(pool_size=(2,2), strides=2, padding='same')(output)
    
    output = Flatten()(output)
    output=Dense(512, activation='relu')(output)
    output=Dense(256, activation='relu')(output)
    output=Dense(128, activation='relu')(output)
    
    output=Dense(2, activation='softmax')(output)
    
    model = Model(inputs=[input], outputs=output)
    model.compile(optimizer='adam',
                 loss='categorical_crossentropy',
                 metrics=['acc'])
    return model
```

```python
CNN_model = CNN_model_build()
CNN_model.summary()
```

```python
history = CNN_model.fit(train_cqt_array, train_labels_cat, epochs=30, batch_size=24, validation_split=0.2)
```

```python
plot_history(history.history)
```

```python
CNN_model.evaluate(test_cqt_array, test_labels_cat)
```

```python
def plot_history(history_dict):
    loss = history_dict['loss']
    val_loss = history_dict['val_loss']
    
    epochs = range(1, len(loss)+1)
    fig = plt.figure(figsize=(14, 5))
    ax1 = fig.add_subplot(1,2,1)
    ax1.plot(epochs, loss, 'b--', label='train_loss')
    ax1.plot(epochs, val_loss, 'r', label='val_loss')
    ax1.set_xlabel("Epochs")
    ax1.set_ylabel("Loss")
    ax1.grid()
    ax1.legend()
    
    acc = history_dict['acc']
    val_acc = history_dict['val_acc']
    ax2 = fig.add_subplot(1,2,2)
    ax2.plot(epochs, acc, 'b--', label='train_accuracy')
    ax2.plot(epochs, val_acc, 'r', label='val_accuracy')
    ax2.set_xlabel("Epochs")
    ax2.set_ylabel("accuarcy")
    ax2.grid()
    ax2.legend()
    
    plt.show()
```

## CNN 모델

```python
from keras.utils import to_categorical

train_mfcc_np = np.array(train_mfcc, np.float32)
#train_cqt_np = train_cqt_np.reshape((80, 168*16))
test_mfcc_np = np.array(test_mfcc, np.float32)
#test_cqt_np = test_cqt_np.reshape((20, 168*16))

train_mfcc_array = np.expand_dims(train_mfcc_np, -1)
test_mfcc_array = np.expand_dims(test_mfcc_np, -1)

train_labels_cat = to_categorical(train_labels_np)
test_labels_cat = to_categorical(test_labels_np)

print(train_mfcc_array.shape)
print(train_labels_cat.shape)
print(test_mfcc_array.shape)
print(test_labels_cat.shape)
```

```python
from keras.layers import Conv2D, MaxPool2D, Flatten

def CNN_model_build():
    model = Sequential()
    
    input = Input(shape=(20, 32, 1))
    output = Conv2D(128, 3, strides=1, padding='same', activation='relu')(input)
    output = MaxPool2D(pool_size=(2,2), strides=2, padding='same')(output)
    
    output = Conv2D(256, 3, strides=1, padding='same', activation='relu')(output)
    output = MaxPool2D(pool_size=(2,2), strides=2, padding='same')(output)
    
    output = Conv2D(512, 3, strides=1, padding='same', activation='relu')(output)
    output = MaxPool2D(pool_size=(2,2), strides=2, padding='same')(output)
    
    output = Flatten()(output)
    output=Dense(512, activation='relu')(output)
    output=Dense(256, activation='relu')(output)
    output=Dense(128, activation='relu')(output)
    
    output=Dense(2, activation='softmax')(output)
    
    model = Model(inputs=[input], outputs=output)
    model.compile(optimizer='adam',
                 loss='categorical_crossentropy',
                 metrics=['acc'])
    return model
```

```python
CNN_model = CNN_model_build()
CNN_model.summary()
```

```python
history = CNN_model.fit(train_mfcc_array, train_labels_cat, epochs=30, batch_size=24, validation_split=0.2)
```

```python
plot_history(history.history)
```

```python
CNN_model.evaluate(test_mfcc_array, test_labels_cat)
```

```python
pred = CNN_model(test_mfcc_array)
## classification_report
from sklearn.metrics import classification_report
y_test = test_labels_np
preds_list = pred.numpy().argmax(-1)
clf_report = classification_report(y_test, preds_list)
print(clf_report)

# 오차행렬 생성
from sklearn.metrics import classification_report, confusion_matrix
cf_matrix = confusion_matrix(y_test, preds_list)
# 오차행렬 시각화
import matplotlib.pyplot as plt
import seaborn as sns
sns.heatmap(cf_matrix, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predict')
plt.ylabel('Actual')
plt.show()
```
